```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
#plot a scatter plot between total_bill and tips
```

```python
#plot here 
sns.relplot(x = 'total_bill' , y = 'tip', kind = 'scatter' , data = data) #in this scatter is default 
```

```python
sns.relplot(x = 'total_bill' , y = 'tip', kind = 'line' , data = data) #you can change kind scatter to line
```

```python
#hue prameter
```

```python
sns.relplot(x = 'total_bill' , y = 'tip', hue = 'smoker' , data = data) #hue make huge diffrent 
```

```python
#style prameter

sns.relplot(x = 'total_bill' , y = 'tip', kind = 'scatter', hue ='smoker', style = 'sex' , data = data)
```

```python
#size prameter
```

```python
sns.relplot(x = 'total_bill' , y = 'tip', kind = 'scatter', hue ='smoker', size = 'size', style = 'sex' , data = data)
```

# using scatter plot

```python
sns.scatterplot(x = 'total_bill' , y = 'tip', hue ='smoker', size = 'size', style = 'sex' , data = data)
```

# Heatmap

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
x = df.pivot_table(index = 'year' , columns ='month' , values ='passengers' , aggfunc ='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x , linewidths = 0.5)
```

```python
sns.heatmap(x , linewidths = 0.5 , annot = True , fmt ='d')
```

```python
sns.heatmap(x , linewidths = 0.5 , annot = True , fmt ='d' , cmap ='viridis')
```

```python
sns.heatmap(x , linewidths = 0.5 , annot = True , fmt ='d' , cmap ='summer')
```

```python
sns.heatmap(x , linewidths = 0.5 , annot = True , fmt ='d' , cmap ='winter')
```

# Clustermap

```python
sns.clustermap(x) #its has cluster similar rows and columns and its default is equl-distance
```

```python
sns.clustermap(x , metric = 'correlation')
```

```python
sns.clustermap(x ,z_score=0, metric = 'correlation') #standarzide
```

```python
sns.clustermap(x , z_score = 0, annot = True , metric = 'correlation')
```

```python
sns.clustermap(x , z_score = 0,row_cluster = False ,annot = True , metric = 'correlation') #you can alson do columns culster
```

# Distplots

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean() , inplace = True)
```

```python
#plotting distplots
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'] , kde = False , bins = 10) #only giving histogram you can also change bins 
```

```python
#plotting only kde not a histogram

sns.distplot(titanic['age'] , hist = False)
```

```python
#plotting only rugplot
```

```python
sns.distplot(titanic['age'] , kde = False , hist =False , rug = True)
```

```python
#how to use distplot
```

```python
sns.distplot(titanic[titanic['survived']==0]['age'])
sns.distplot(titanic[titanic['survived']==1]['age'])
```
